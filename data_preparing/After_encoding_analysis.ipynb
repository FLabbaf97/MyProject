```python
save_processed_path = '../data/processed/'
expression_filtered_norm = pd.read_csv(
    save_processed_path+'DepMap_expression_processed_1383_15806.csv')
```

```python
import plotly.express as px
from MulticoreTSNE import MulticoreTSNE as TSNE
tsne = TSNE(n_jobs=4)
```

# before encoding

```python
x = expression_filtered_norm.iloc[:, 1:-3].values
Y=tsne.fit_transform(x)
Y.shape
dft = pd.DataFrame()
dft['x']=Y[:,0]
dft['y']=Y[:,1]
dft['tissue'] = expression_filtered_norm.tissue.values
dft['disease']=expression_filtered_norm.disease.values

px.scatter(dft,x='x',y='y',color='disease').show()
px.scatter(dft,x='x',y='y',color='tissue').show()
```
