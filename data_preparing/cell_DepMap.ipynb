# preprocessing data

```python
# gene expression
expression = pd.read_csv(save_path+"DepMap_expression.csv")
expression.set_index('cell_line_name' , inplace=True)
expression
```

```python
expression = expression.fillna(0)
expression = expression[expression.columns[(expression == 0).mean() < 0.50]]
expression
```

```python
# remove low variance genes
from sklearn.feature_selection import VarianceThreshold
# Removing both constant and quasi-constant
var_thr = VarianceThreshold(threshold=0.1)
var_thr.fit(expression)
var_thr.get_support()
concol = [column for column in expression.columns 
                       if column not in expression.columns[var_thr.get_support()]]
expression_filtered = expression.drop(concol, axis=1)
expression_filtered
```

```python
import matplotlib.pyplot as plt

stds = expression[expression.columns].std()
std2 = expression_filtered[expression_filtered.columns].std()
mean1 = expression[expression.columns].mean()
mean2 = expression_filtered[expression_filtered.columns].mean()
x=plt.hist(stds, bins=100)
plt.show()
x=plt.hist(mean1, bins=100)
plt.show()
x=plt.hist(std2, bins=100)
plt.show()
x=plt.hist(mean2, bins=100)
plt.show()
```

## Min-Max normalization

```python
from re import M
from matplotlib.pyplot import axis


a = expression_filtered.values
a.shape
# a
# mean = a.mean(axis=0,keepdims=True)
# mean
# mean.shape
# s = a.std(axis=0, keepdims=True)
# s
m = a.min(axis = 0,keepdims=True)
m
M = a.max(axis = 0,keepdims=True)
M
n = ((a-m)/(M-m))
n
expression_filtered_norm = expression_filtered
expression_filtered_norm.iloc[:,:]=n
expression_filtered_norm
```

```python
# Add metadata to dataframe
meta = pd.read_csv(save_path+'DepMap_meta.csv')
meta
```

```python

expression_filtered_norm.reset_index(inplace=True)
# expression_filtered_norm.rename({'cell_line_name':'model_id'}, inplace = True)
expression_filtered_norm['tissue'] = expression_filtered_norm['cell_line_name'].apply(
    lambda x: meta[meta['cell_line_name'] == x]['lineage'].iloc[0])
expression_filtered_norm['cancer_type'] = expression_filtered_norm['cell_line_name'].apply(
    lambda x: meta[meta['cell_line_name'] == x]['primary_disease'].iloc[0])
expression_filtered_norm['disease'] = expression_filtered_norm['cancer_type'].apply(
    lambda x: x.split(' ')[-1].lower() if type(x) is not float else '')
# cell_line_meta_broad
expression_filtered_norm.set_index('cell_line_name', inplace=True)
expression_filtered_norm
```

```python
#save
expression_filtered_norm.to_csv(save_path+'DepMap_expression_processed.csv')
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
expression_filtered_norm_train, expression_filtered_norm_test = train_test_split(
    expression_filtered_norm, shuffle=True, test_size=0.2)
len(expression_filtered_norm_train)
len(expression_filtered_norm_test)
```
